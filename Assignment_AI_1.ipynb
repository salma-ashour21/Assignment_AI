```python
import numpy as np
import pandas as pd

data = pd.read_csv("/content/student_scores.csv")
X = data["Hours"].values
y = data["Scores"].values

w = 0
b = 0
learning_rate = 0.001
epochs = 200
loss_history = []

#y=w*x=b
for epoch in range(epochs):
    for i in range(len(X)):
        xi = X[i]
        yi = y[i]

        y_pred = w * xi + b
        error = y_pred - yi

        w = w - learning_rate * error * xi
        b = b - learning_rate * error
    total_loss = np.mean((w * X + b - y) ** 2)
    loss_history.append(total_loss)

    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch+1}: Loss = {total_loss:.3f}")

print(f"Final weight: {w:.3f}, bias: {b:.3f}")

#test
new_x = 5
predicted_y = w * new_x + b
print(f"Predicted value for x={new_x}: {predicted_y:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='red', label='scores')
plt.plot(X, w * X + b, color='blue', label='scores_pre')
plt.xlabel("hours")
plt.ylabel("y_pre")
plt.title("Linear Regression used SGD")
plt.legend()
plt.show()
```

```python
plt.plot(loss_history)
plt.title("Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
